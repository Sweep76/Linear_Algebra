```python
# 1. Let u = (1,-2,4) and v = (3,5,1) w = (2,1,-3)
import numpy as np
u = np.array([1,-2,4])
v = np.array([3,5,1])
w = np.array([2,1,-3])

# a. 3u - 2v
# b. 5u + 3v - 4w
print("1. Let u = (1,-2,4) and v = (3,5,1) w = (2,1,-3)\n")
print("a. 3u-2v:",(3*u)-(2*v))
print("b. 5u+3v-4w:", (5*u) + (3*v) - (4*w))

# c. u.v, v.w, u.w
print("c. u.v, v.w, u.w")
print("  u.v = ", u.dot(v))
print("  v.w = ", v.dot(w))
print("  u.w = ", u.dot(w))

# d. ||u||,||v||,||w||
# Let u = (1,-2,4) and v = (3,5,1) w = (2,1,-3)
print("\nd. ||u||,||v||,||w||")
print("  ||u||: ", np.sqrt(u.dot(u)))
print("  ||v||: ", np.sqrt(v.dot(v)))
print("  ||w||: ", np.sqrt(w.dot(w)))

#test

# e. if theta is the angle between u and v find cos 0
print("\ne. if theta is the angle between u and v find cos 0")
print("angle:",u.dot(v)/(np.linalg.norm(u)*np.linalg.norm(v)))

# f. if a is the angle between v and w find cos a
print("\nf. if a is the angle between v and w find cos a")
print("angle:",v.dot(w)/(np.linalg.norm(v)*np.linalg.norm(w)))

# g. if beta is the angle between u and w find cos b
print("\ng. if beta is the angle between u and w find cos b")
print("angle:",u.dot(w)/(np.linalg.norm(u)*np.linalg.norm(w)))




```

```python
# 1 part II. Let u = (1,-2,4) and v = (3,5,1) w = (2,1,-3)
import numpy as np
u = np.array([1,-2,4])
v = np.array([3,5,1])
w = np.array([2,1,-3])

# h. Find d(u,v), Let u = (1,-2,4) and v = (3,5,1) w = (2,1,-3)
print("\nh. d(u,v)")
print("d(u,v): ", np.linalg.norm(u-v))


# i. Find d(w,v)
print("\ni. d(w,v)")
print("d(w,v): ", np.linalg.norm(w-v))

# j. Find d(u,w)
print("\nj. d(u,w)")
print("d(u,w): ", np.linalg.norm(u-w))

# k. Find proj(u, v)
print("\nk. Find proj(u, v)")
print("  proj(u,v): ",u.dot(v)*v/(np.linalg.norm(v)**2) )

# l. Find proj(w, v)
print("\nl. Find proj(w, v)")
print("  proj(w,v): ",w.dot(v)*v/(np.linalg.norm(v)**2) )

# m. Find proj(u, w)
print("\nm. Find proj(u, w)")
print("  proj(u,v): ",u.dot(w)*w/(np.linalg.norm(w)**2) )
```
